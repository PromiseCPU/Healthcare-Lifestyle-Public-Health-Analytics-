### Targeted Intervention Filtering (Identifying High-Risk Individuals)

#### Screening for Cardiac Risk by Boolean Indexing

```python
# Identifying patients that need immediate attention 
# Using 'High Risk' as a measure if they Smoke, exercise less than 2hrs and possess BMI greater than 30 (Obese) altogether

import pandas as pd                      
import numpy as np                      
import matplotlib.pyplot as plt         
import seaborn as sns

# loading the dataframe
df = pd.read_csv(r'C:\Users\PROMISECPU\Python Public Health\health_activity_data.csv')

high_risk_criteria = (df['Smoker'] == 'Yes') & (df['Exercise_Hours_per_Week'] < 2) & (df['BMI'] > 30)
priority_patients = df[high_risk_criteria].copy()
non_high_risk = df[~ high_risk_criteria]

# Extracting systolic BP for numeric analysis
priority_patients['Systolic_BP'] = priority_patients['Blood_Pressure'].str.split('/').str[0].astype(float)
non_high_risk['Systolic_BP'] = non_high_risk['Blood_Pressure'].str.split('/').str[0].astype(float)

print(f"High-risk individuals: {len(priority_patients)}")
print(f"Remaining individuals: {len(non_high_risk)}")
```

#### Summary Statistics (Comparing mean and median for key numeric values)

```python
cols = ['Age','Heart_Rate','Systolic_BP', 'BMI','Exercise_Hours_per_Week']
print('=== High-Risk Group ===')
print(priority_patients[cols].describe())

print('\n=== Non High-Risk Group ===')
print(non_high_risk[cols].describe())
```

#### Visual Composition by creating multiple subplots (Boxplots to visually compare distributions)

```python
# Boxplots (Subplots) showing the variation of Age, Heart rate, Systolic BP and BMI against 'High_Risk Group'

import pandas as pd  
import matplotlib.pyplot as plt
import seaborn as sns

# Create combined dataset for comparison
priority_patients['Group'] = 'High-Risk' 
non_high_risk['Group'] = 'Non High-Risk'

# Extract Systolic BP as numeric for both groups (handling missing values
priority_patients['Systolic_BP'] = priority_patients['Blood_Pressure'].str.split('/').str[0]
priority_patients['Systolic_BP'] = pd.to_numeric(priority_patients['Systolic_BP'], errors = 'coerce')

non_high_risk['Systolic_BP'] = non_high_risk['Blood_Pressure'].str.split('/').str[0]
non_high_risk['Systolic_BP'] = pd.to_numeric(non_high_risk['Systolic_BP'],  errors = 'coerce')

# Combine into one Dataframe for plotting
combined_df = pd.concat([priority_patients, non_high_risk])

# Plot boxplots for key numeric with different colors
fig, axes = plt.subplots(2,2, figsize=(12,10))
axes = axes.flatten()
variables = ['Age', 'Heart_Rate','Systolic_BP', 'BMI']

for i, var in enumerate(variables):
    sns.boxplot(x = 'Group', y = var, hue = 'Group', data = combined_df, palette = 'Set2', legend = False, ax = axes[i])
    axes[i].set_title(f'{var} by Risk Group')
    axes[i].set_xlabel('Risk Group')
    axes[i].set_ylabel(var)

plt.tight_layout()
plt.show()
```

#### Correlation within High-Risk Group using HeatMap (Multivariate Analysis)

```python
import pandas as pd  
import matplotlib.pyplot as plt
import seaborn as sns

corr_matrix = priority_patients[cols].corr()
sns.heatmap(corr_matrix, annot = True, cmap = 'PuOr', center = 0)
plt.title('Correlation within High-Risk Group')
plt.show()
```

```python
# +1: Perfect Positive Relationship (as one variable increases, the other increases and vice-versa)
# 0: No consistent relationship
# -1: Perfect Negative Relationship (as one variable increases, the other decreases and vice-versa)

# Stronger positive correlations appear in darker warm colours, Stronger negative correlations appear in cooler colours

# From the above heatmap, 
# Age - Heart_rate (0.44) is a moderate positive correlation, that is, older patients tend to have higher heart rates
# Age - Exercise_Hours_per_Week (-0.58) is a moderately strong negative correlation where older patients tend to exercise less and so on
```

#### Correlation Matrix Visualization using Heatmap (Multivariate Analysis)

```python
# Instead of running 20+ different charts, this multivariate image (Heatmap) gives you a complete overview of variable relationships 

import pandas as pd  
import numpy as np                       
import matplotlib.pyplot as plt          
import seaborn as sns  

df = pd.read_csv(r'C:\Users\PROMISECPU\Python Public Health\health_activity_data.csv')
numeric_df = df.select_dtypes(include = [np.number])
correlation_matrix = numeric_df.corr()

# Set up the figure size
plt.figure(figsize = (12, 10))

# Create the heatmap using Seaborn
sns.heatmap(correlation_matrix, annot = True, cmap = 'Spectral', fmt = '.2f', linewidths = 0.5, center = 0)

# Add title and Display
plt.title('Public Health Heatmap: Variable Inter-correlations', fontsize = 16)
plt.show()
```
